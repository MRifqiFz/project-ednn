```python
import os
import pandas as pd
```

```python
listdir = os.listdir('./Metrics')
listdir
```

```python
a = []
b = []
for i in listdir:
    if i[0] != '[':
        x = pd.read_csv('./Metrics/' + i)
        a.append(x)
    else:
        x = pd.read_csv('./Metrics/' + i)
        b.append(x)
```

```python
b
```

```python
results1 = pd.concat(a)
for i in range(len(results1)):
    original_string = results1['Precision (Exp)'].iloc[i]
    start_index = original_string.find('(') + 1
    end_index = original_string.find(',')
    numerical_value = float(original_string[start_index:end_index])
    results1['Precision (Exp)'].iloc[i] = numerical_value

for i in range(len(results1)):
    original_string = results1['Recall (Exp)'].iloc[i]
    original_string2 = results1['Accuracy (Exp)'].iloc[i]
    start_index = original_string.find('(') + 1
    end_index = original_string.find(')')
    numerical_value = float(original_string[start_index:end_index])
    numerical_value2 = float(original_string2[start_index:end_index])
    results1['Recall (Exp)'].iloc[i] = numerical_value
    results1['Accuracy (Exp)'].iloc[i] = numerical_value2

results1

```

```python
results2 = pd.concat(b)
for i in range(len(results2)):
    original_string = results2['Precision (Exp)'].iloc[i]
    start_index = original_string.find('(') + 1
    end_index = original_string.find(',')
    numerical_value = float(original_string[start_index:end_index])
    results2['Precision (Exp)'].iloc[i] = numerical_value

for i in range(len(results2)):
    original_string = results2['Recall (Exp)'].iloc[i]
    original_string2 = results2['Accuracy (Exp)'].iloc[i]
    start_index = original_string.find('(') + 1
    end_index = original_string.find(')')
    numerical_value = float(original_string[start_index:end_index])
    numerical_value2 = float(original_string2[start_index:end_index])
    results2['Recall (Exp)'].iloc[i] = numerical_value
    results2['Accuracy (Exp)'].iloc[i] = numerical_value2

results2
```

```python
results1.to_csv('results1.csv')
results2.to_csv('results2.csv')
```
