```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import textwrap
```

```python
excel_path = "MLC Metrics - Separated.xlsx"
```

```python
example_based = pd.read_excel(excel_path, sheet_name='Example Based')
example_based
```

```python
label_based = pd.read_excel(excel_path, sheet_name='Label Based')
label_based
```

```python
def wrap_labels(ax, width):
    labels = [textwrap.fill(label.get_text(), width) for label in ax.get_xticklabels()]
    ax.set_xticklabels(labels, rotation=0)


for i in range(10):
    g = sns.catplot(
        data=example_based,  kind="bar", x="MLC Framework", y=example_based.columns[2:][i], hue="Classifier", palette="dark", alpha=.8, height=4, aspect=2
    )
    wrap_labels(g.ax, 10)   
```
