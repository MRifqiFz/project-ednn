```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import textwrap
```

```python
excel_path = "MLC Metrics - Separated.xlsx"
```

```python
example_based = pd.read_excel(excel_path, sheet_name='Example Based')
example_based
```

```python
label_based = pd.read_excel(excel_path, sheet_name='Label Based')
label_based
```

```python
sns.color_palette("bright6", 10)
```

```python
def wrap_labels(ax, width):
    labels = [textwrap.fill(label.get_text(), width) for label in ax.get_xticklabels()]
    ax.set_xticklabels(labels, rotation=0)


for i in range(10):
    g = sns.catplot(
        data=example_based,  kind="bar", x="MLC Framework", y=example_based.columns[2:][i], hue="Classifier", palette="bright", alpha=.8, height=5, aspect=2
    )
    wrap_labels(g.ax, 10)
    plt.savefig(f"Visualization/Example/{example_based.columns[2:][i]}.png")
```

```python
for i in range(10):
    g = sns.catplot(
        data=label_based,  kind="bar", x="MLC Framework", y=label_based.columns[2:][i], hue="Classifier", palette="bright", alpha=.8, height=5, aspect=2
    )
    wrap_labels(g.ax, 10)
    plt.savefig(f"Visualization/Label/{label_based.columns[2:][i]}.png")
```

```python
from PIL import Image

def create_collage(width, height, listofimages, cols, rows, filename):
    thumbnail_width = width//cols
    thumbnail_height = height//rows
    size = thumbnail_width, thumbnail_height
    new_im = Image.new('RGB', (width, height))
    ims = []
    for p in listofimages:
        im = Image.open(p)
        im.thumbnail(size)
        ims.append(im)
    i = 0
    x = 0
    y = 0
    for col in range(cols):
        for row in range(rows):
            print(i, x, y)
            new_im.paste(ims[i], (x, y))
            i += 1
            y += thumbnail_height
        x += thumbnail_width
        y = 0
    new_im.save(filename)
```

```python
import os
example_path = "Visualization/Example/"
example_image = [example_path+i for i in os.listdir(example_path)]
example_image

label_path = "Visualization/Label/"
label_image = [label_path+i for i in os.listdir(label_path)]
```

```python
create_collage(5000, 840, example_image, 5, 2, "example.png")
```

```python
create_collage(5000, 840, label_image, 5, 2, "label.png")
```
